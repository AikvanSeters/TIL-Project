```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#Data_Sets
#Read Data_sets
file_path1 = 'Data/COVID-19_ziekenhuisopnames_1of2.csv'
covid_temp1 = pd.read_csv(file_path1, delimiter = ';',  parse_dates=['Date_of_statistics'], index_col = 'Date_of_statistics')

file_path2 = 'Data/COVID-19_ziekenhuisopnames_2of2.csv'
covid_temp2 = pd.read_csv(file_path2, delimiter = ';',  parse_dates=['Date_of_statistics'], index_col = 'Date_of_statistics')

    #Merge both Data_sets
covid = pd.merge(covid_temp1, covid_temp2, how = 'outer')
covid
# covid.Hospital_admission.resample('M' ,kind='period').sum()
#for i in covid['Date_of_statistics']:
 #   print(i.month)
```

```python
covid_province = covid.groupby(['Date_of_statistics', 'Security_region_name']).sum()
covid_province.reset_index(inplace=True)
#covid_province['Datetime'] = pd.to_datetime(covid_province['Date_of_publication'])
fig = px.scatter(covid_province, x = 'Date_of_statistics',  y = 'Hospital_admission', color = 'Security_region_name')
fig.update_yaxes(range = [0, 100])
```
